```python
# rescale / normalize data between 0 and 1
"""
Scales features to a fixed range, typically [0, 1] or [-1, 1].
Formula: X_scaled = (X - X_min) / (X_max - X_min)
Use when: You want to preserve zero entries in sparse data or when you need exact bounds on your values.
Pros: Preserves zero entries, doesn't change the distribution shape.
Cons: Doesn't reduce the impact of outliers.
"""

from sklearn.preprocessing import MinMaxScaler

array = mydataframe.values

# seaparate input and output variables
X = array[:, 0:8]
Y = array[:, 8]

scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(X)
print(rescaledX)
```

```python
# standardize data
"""
Standardizes features by removing the mean and scaling to unit variance.
Formula: X_scaled = (X - mean(X)) / std(X)
Use when: You assume your data follows a Gaussian distribution and when you want to handle outliers.
Pros: Handles outliers well, centers the data around zero.
Cons: Doesn't produce normalized data with a fixed range.
"""

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X)
rescaledX = scaler.transform(X)
print(rescaledX)
```

```python
# normalize data
"""
Scales each sample (row) to have unit norm (length 1).
Formula depends on the chosen norm (l1, l2, or max).
Use when: You want to scale samples independently of each other, often in text classification or clustering.
Pros: Useful for sparse datasets, helps in comparing features on different scales.
Cons: Loses absolute scale information between samples.
"""

from sklearn.preprocessing import Normalizer

scaler = Normalizer().fit(X)
normalizedX = scaler.transform(X)
print(normalizedX)
```

```python
# binarize data
"""
Transforms features to binary values based on a threshold.
Formula: X_binary = 1 if X > threshold else 0
Use when: You need to convert numerical features to boolean features.
Pros: Simplifies data, useful for algorithms that work with binary inputs.
Cons: Loses granularity in the data.
"""

from sklearn.preprocessing import Binarizer

binarizer = Binarizer(threshold=0.0).fit(X)
binaryX = binarizer.transform(X)
print(binaryX)
```

# Feature Selection for Machine Learning

Process to select features in data that contribute most to the prediction variable or output, having irrelevant features can decrease the accuracy of many models. 

Performing feature selection before modelling can reduce overfitting (redundant features creating noise), improve accuracy and reduce training time. 

- univariate selection
- recursive feature elimination
- principle component analysis
- feature importance

```python
# univariate selection
import numpy as np
from sklearn.feature_selection import SelectKBest, chi2

feature_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age']

# feature extraction, select 4 of the best features using chi-squared test for non-negative features
test = SelectKBest(score_func=chi2, k=4)
fit = test.fit(X, Y)

# Get the indices of the selected features
selected_feature_indices = test.get_support(indices=True)

# Get the names of the selected features
selected_feature_names = [feature_names[i] for i in selected_feature_indices]

# Print the selected feature names
print("Selected features:")
for name, score in zip(selected_feature_names, fit.scores_[selected_feature_indices]):
    print(f"{name}: {score:.3f}")

# summarize scores
np.set_printoptions(precision=3)
print("\nAll scores:")
print(fit.scores_)

# create a new dataset that only includes these selected features
# reduce the dimensionality 
features = fit.transform(X)
print("\nTransformed features (first 5 rows):")
print(features[0:5, :])
```

```python
# recursive feature elimination

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import numpy as np

# Assuming these are the feature names in your dataset
feature_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age']

values = mydataframe.values
X = values[:, 0:8]
Y = values[:, 8]

model = LogisticRegression()
rfe = RFE(estimator=model, n_features_to_select=3)
fit = rfe.fit(X, Y)

print("Num Features: %d" % fit.n_features_to_select)

# Get the mask of selected features
selected_features_mask = fit.support_

# Get the names of selected features
selected_features = np.array(feature_names)[selected_features_mask]

print("Selected Features:")
for feature, rank in zip(feature_names, fit.ranking_):
    if feature in selected_features:
        print(f"{feature}: selected (rank: {rank})")
    else:
        print(f"{feature}: not selected (rank: {rank})")

print("\nFeature Ranking:")
for feature, rank in sorted(zip(feature_names, fit.ranking_), key=lambda x: x[1]):
    print(f"{feature}: {rank}")
```

```python
# principal component analysis

from sklearn.decomposition import PCA
import pandas as pd

values = mydataframe.values
X = values[:, 0:8]
Y = values[:, 8]

# Add the feature names
feature_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age']

# Create the PCA model
pca = PCA(n_components=3)
fit = pca.fit(X)

# Summarize components
print("Explained Variance: %s" % fit.explained_variance_ratio_)
print(fit.components_)

# Create a DataFrame to display feature names with their loadings
pca_components = pd.DataFrame(
    fit.components_.T,  # Transpose the components matrix
    columns=[f'PC{i+1}' for i in range(fit.n_components_)],
    index=feature_names
)

print("\nPCA Components:")
print(pca_components)

# Print top features for each principal component
for i, pc in enumerate(pca_components.columns):
    print(f"\nTop features in {pc}:")
    print(pca_components[pc].abs().sort_values(ascending=False))
```

```python
# feature importance

from sklearn.ensemble import ExtraTreesClassifier

values = mydataframe.values
X = values[:, 0:8]
Y = values[:, 8]

# feature names
feature_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age']

model = ExtraTreesClassifier()
model.fit(X, Y)

# get feature importances and pair them with feature names
importances = model.feature_importances_
feature_importances = list(zip(feature_names, importances))

# sort feature importances in descending order
feature_importances = sorted(feature_importances, key=lambda x: x[1], reverse=True)

# print sorted feature importances
for name, value in feature_importances:
    print(f"{name}: {value:.3f}")

# identify the feature with the highest importance
highest_importance = feature_importances[0][0]
print(f"Feature with the highest importance: {highest_importance}")
```

# Evaluate machine learning algorithms with resampling

- train and test sets
- k-fold cross validation
- leave one out cross validation
- repeated random test-train splits

```python
# train and test sets

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

# load dataset
filename = 'datasets/diabetes.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = pandas.read_csv(filename, skiprows=1, names=names)

array = dataframe.values
X = array[:, 0:8]
Y = array[:, 8]

# split into train and test sets
test_size = 0.33
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)

# fit model
model = LogisticRegression()
model.fit(X_train, Y_train)
result = model.score(X_test, Y_test)
print("Accuracy without any preprocessing or feature selection: %.3f%%" % (result * 100.0))

# another model with the transformed data, and feature selection
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(X)

pca = PCA(n_components=6)
selected_features = pca.fit_transform(rescaledX)

X_train, X_test, Y_train, Y_test = train_test_split(selected_features, Y, test_size=test_size, random_state=seed)

# train and evaluate the model
model = LogisticRegression()
model.fit(X_train, Y_train)
result = model.score(X_test, Y_test)
print("Accuracy with preprocessing and feature selection: %.3f%%" % (result * 100.0))
```

```python
# k-fold cross validation

# algo is trained and evaluated multiple times on different data.

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

values = dataframe.values
X = values[:, 0:8]
Y = values[:, 8]

num_folds = 10
seed = 7
kfold = KFold(n_splits=num_folds, shuffle=True, random_state=seed)
model = LogisticRegression()
results = cross_val_score(model, X, Y, cv=kfold)
print("Accuracy: %.3f%% (%.3f%%)" % (results.mean()*100.0, results.std()*100.0))

```

```python
# random test-train splits

from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

n_splits = 10
test_size = 0.33
seed = 7
kfold = ShuffleSplit(n_splits=n_splits, test_size=test_size, random_state=seed)
model = LogisticRegression()
results = cross_val_score(model, X, Y, cv=kfold)
print("Accuracy: %.3f%% (%.3f%%)" % (results.mean()*100.0, results.std()*100.0))
```

```python
# logarithmic loss

"""
For binary classification:
Log Loss = -1/N Σ(y_i log(p_i) + (1 - y_i) log(1 - p_i))
Where:
N is the number of samples
y_i is the true label (0 or 1)
p_i is the predicted probability of the positive class
"""

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

values = dataframe.values
X = values[:, 0:8]
Y = values[:, 8]

num_folds = 10
seed = 7
kfold = KFold(n_splits=num_folds, shuffle=True, random_state=seed)
model = LogisticRegression()
scoring = 'neg_log_loss'
results = cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
print("Accuracy: %.3f%% (%.3f%%)" % (results.mean()*100.0, results.std()*100.0))
```

# Automate Machine Learning Workflows with Pipelines

- use pipelines to minimize data leakage, ensure sequence of data transforms to be chained together and the pipelines are constrained to the data available for the evaluation
- construct a data preparation and modeling pipeline
- construct a feature extraction and modeling pipeline
  


```python
# data preparation and modeling pipeline

# separate training and testing before preprocessing to prevent training data biased by the testing data (influenced by the scale of data in the test set)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

# load data
filename = 'datasets/diabetes.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = pandas.read_csv(filename, names=names, skiprows=1)
array = dataframe.values
X = array[:, 0:8]
Y = array[:, 8]
test_size = 0.33
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)

# create pipeline
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('lda', LinearDiscriminantAnalysis()))
estimators.append(('lr', LogisticRegression()))
model = Pipeline(estimators)

# evaluate pipeline
kfold = KFold(n_splits=10, random_state=7, shuffle=True)
results = cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
# feature extraction and modeling pipeline

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.decomposition import PCA
from sklearn.pipeline import FeatureUnion


filename = 'datasets/diabetes.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = pandas.read_csv(filename, names=names, skiprows=1)
array = dataframe.values
X = array[:, 0:8]
Y = array[:, 8]
test_size = 0.33
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)

# create feature union, allow the result of multiple feature selection and extraction procedures to be combined into a larger dataset
features = FeatureUnion([
    ('pca', PCA(n_components=3)),
    ('select_best', SelectKBest(k=6))
])

# create pipeline
pipeline = Pipeline([
    ('features', features),
    ('logistic', LogisticRegression())
])

# perform cross-validation on the training data
cv_scores = cross_val_score(pipeline, X_train, Y_train, cv=kfold)
print("Cross-validation scores: %.3f (%.3f)" % (cv_scores.mean(), cv_scores.std()))

# fit the pipeline on the entire training data
pipeline.fit(X_train, Y_train)

# evaluate on the test set
test_score = pipeline.score(X_test, Y_test)
print("test set score: %.3f" % test_score)
```

# Improve model performance with ensembles

- bagging: building multiple models from different subsamples of the training dataset
- boosting: building multiple models, each of which learns to fix the prediction errors of a prior model in the sequence of models
- voting: building multiple models, and use simple statistics to combine predictions

```python
# bagging algorithms

from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

filename = 'datasets/diabetes.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = pandas.read_csv(filename, names=names, skiprows=1)
array = dataframe.values
X = array[:, 0:8]
Y = array[:, 8]

kfold = KFold(n_splits=10, random_state=7, shuffle=True)
cart = DecisionTreeClassifier()
num_trees = 100
model = BaggingClassifier(estimator=cart, n_estimators=num_trees, random_state=7)
results = cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
# random forest
'''
Samples of the training dataset are
taken with replacement, but the trees are constructed in a way that reduces the correlation
between individual classifiers. Specifically, rather than greedily choosing the best split point in
the construction of each tree, only a random subset of features are considered for each split.
'''

from sklearn.ensemble import RandomForestClassifier

filename = 'datasets/diabetes.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = pandas.read_csv(filename, names=names, skiprows=1)
array = dataframe.values
X = array[:, 0:8]
Y = array[:, 8]

num_trees = 100
max_features = 3
kfold = KFold(n_splits=10, random_state=7, shuffle=True)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
# adaboost
"""

    An AdaBoost [1]_ classifier is a meta-estimator that begins by fitting a
    classifier on the original dataset and then fits additional copies of the
    classifier on the same dataset but where the weights of incorrectly
    classified instances are adjusted such that subsequent classifiers focus
    more on difficult cases.
"""

from sklearn.ensemble import AdaBoostClassifier

filename = 'datasets/diabetes.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = pandas.read_csv(filename, names=names, skiprows=1)
array = dataframe.values
X = array[:, 0:8]
Y = array[:, 8]

num_trees = 30
seed = 7
kfold = KFold(n_splits=10, random_state=seed, shuffle=True)
model = AdaBoostClassifier(n_estimators=num_trees, algorithm='SAMME', random_state=seed)
results = cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
# voting ensemble

from sklearn.ensemble import VotingClassifier

filename = 'datasets/diabetes.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = pandas.read_csv(filename, names=names, skiprows=1)
array = dataframe.values
X = array[:, 0:8]
Y = array[:, 8]

# create the sub models
estimators = []
model1 = LogisticRegression(max_iter=500)
estimators.append(('logistic', model1))
model2 = DecisionTreeClassifier()
estimators.append(('cart', model2))
model3 = SVC()
estimators.append(('svm', model3))

# create the ensemble model
ensemble = VotingClassifier(estimators)
results = cross_val_score(ensemble, X, Y, cv=kfold)
print(results.mean())
```
